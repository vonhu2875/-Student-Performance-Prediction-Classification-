```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv(r"..\data\student_performance.csv")
df.head()
# Biểu đồ 1: Histogram phân phối điểm môn Toán
plt.figure(figsize=(8,6))
plt.hist(df['math score'], bins=20, edgecolor='black')
plt.title('Phân phối điểm môn Toán (Math Score)')
plt.xlabel('Điểm Toán')
plt.ylabel('Số lượng học sinh')
plt.show()
# Biểu đồ 2: Bar chart so sánh số lượng học sinh Nam và Nữ
plt.figure(figsize=(8,6))
df['gender'].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.title('So sánh số lượng học sinh Nam và Nữ')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng học sinh')
plt.xticks(rotation=0)
plt.show()
# Biểu đồ 3: Boxplot so sánh điểm Toán giữa Nam và Nữ
plt.figure(figsize=(8,6))
df.boxplot(column='math score', by='gender', grid=False)
plt.title('So sánh điểm Toán giữa Nam và Nữ')
plt.suptitle('')  # Remove automatic title
plt.xlabel('Giới tính')
plt.ylabel('Điểm Toán')
plt.show()
```

```python
# 4. Ma tran tuong quan
cols = ['AttendanceRate', 'StudyHoursPerWeek', 'PreviousGrade', 'FinalGrade']
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm')
plt.title('Ma tran tuong quan diem so')
plt.show()

# 5. Tim Outlier theo ho tro tu cha me
sns.boxplot(x='ParentalSupport', y='FinalGrade', data=df)
plt.title('Tim diem di biet theo Parental Support')
plt.show()

# 6. Tuong quan diem qua khu va hien tai
sns.scatterplot(x='PreviousGrade', y='FinalGrade', data=df)
plt.title('Moi quan he diem Cu va Moi')
plt.show()
```
