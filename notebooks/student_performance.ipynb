```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(r"../data/student_performance.csv")
df.head()
```

```python
# Biểu đồ 1: Histogram phân phối điểm cuối kì
plt.figure(figsize=(8,6))
plt.hist(df['FinalGrade'], bins=20, edgecolor='black')
plt.title('Phân phối điểm cuối kì')
plt.xlabel('Điểm cuối kì')
plt.ylabel('Số lượng học sinh')
plt.show()
```

```python
# Biểu đồ 2: Bar chart so sánh số lượng học sinh Nam và Nữ
plt.figure(figsize=(8,6))
df['Gender'].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.title('So sánh số lượng học sinh Nam và Nữ')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng học sinh')
plt.xticks(rotation=0)
plt.show()
```

```python
# Biểu đồ 3: Boxplot so sánh điểm cuối kì giữa Nam và Nữ
plt.figure(figsize=(8,6))
df.boxplot(column='FinalGrade', by='Gender', grid=False)
plt.title('So sánh điểm cuối kì giữa Nam và Nữ')
plt.suptitle('')  # Remove automatic title
plt.xlabel('Giới tính')
plt.ylabel('Điểm cuối kì')
plt.show()
```

```python
# 7. Ảnh hưởng của ngoại khóa
sns.barplot(x='ExtracurricularActivities', y='FinalGrade', data=df)
plt.title('Điểm trung bình hoạt động theo ngoại khóa')
plt.show()
```

```python
# 7. Ảnh hưởng của ngoại khóa làm rõ hơn ở phần đỉnh
sns.barplot(x='ExtracurricularActivities', y='FinalGrade', data=df)
plt.ylim(70, 85)
plt.title('Điểm trung bình hoạt động theo ngoại khóa')
plt.show()
```

```python
# 8. Ảnh hưởng của lớp học online
df['Online Classes Taken'] = df['Online Classes Taken'].map({True: 'Có học online', False: 'Không học online'})
# Sau đó vẽ lại biểu đồ
sns.violinplot(x='Online Classes Taken', y='FinalGrade', data=df)
plt.title('Phân bổ điểm theo lớp học online')
plt.show()
```
